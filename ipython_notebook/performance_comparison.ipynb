```python
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
def get_float_array_from_file(path):
    """
    This function takes a path to a file containing a tab-delimited
    text file and converts that file into an array.

    :param path: Path to the file
    :return: The array
    """
    f = open(path)
    for line in f:
        return map(float, line.split('\t'))
        break
    f.close()
        
def get_results(path):
    
    f = open(path)
    for (i, line) in enumerate(f):
        line = line.rstrip()
        dollars = map(float, line.split('\t'))      
        return dollars
        """
        if i==0:
            sharpe_ratio = line.split(' ')[-1]
        else:
            print 'a'
            dollars = map(float, line.split('\t'))      
            return sharpe_ratio, dollars
        """
        break
        
```

```python
# RMR
rmr_sharpe = 0.0
rmr_dollars_over_time = get_results('../results/new/rmr.txt')
plt.plot(rmr_dollars_over_time)
plt.title('RMR (Sharpe: ' + str(rmr_sharpe) + ')')
plt.ylabel('Dollar Value of Assets')
plt.xlabel('Days')
plt.show()
```

```python
# Uniform Constant Rebalancing
ucr_dollars_over_time = get_float_array_from_file('../results/newconst_rebalancing_dollars_over_time.txt')
plt.plot(ucr_dollars_over_time)
plt.ylabel('Dollar Value of Assets')
plt.xlabel('Days')
plt.show()
```

```python
# Uniform Buy and Hold
ubah_dollars_over_time = get_float_array_from_file('../results/uniform_buy_and_hold_dollars_over_time.txt')
plt.plot(ubah_dollars_over_time)
plt.ylabel('Dollar Value of Assets')
plt.xlabel('Days')
plt.show()
```

```python
# Semi-Constant Uniform Rebalancing Portfolio

# Rebalance every other day
scurp_2_dollars_over_time = get_float_array_from_file('../results/semi_const_rebalancing_dollars_over_time_interval_2.txt')

# Rebalance every 4th day
scurp_4_dollars_over_time = get_float_array_from_file('../results/semi_const_rebalancing_dollars_over_time_interval_4.txt')

# Rebalance every 6th day
scurp_6_dollars_over_time = get_float_array_from_file('../results/semi_const_rebalancing_dollars_over_time_interval_6.txt')


plt.plot(scurp_2_dollars_over_time, label='SCURP2')
plt.plot(scurp_4_dollars_over_time, label='SCURP4')
plt.plot(scurp_6_dollars_over_time, label='SCURP6')
plt.ylabel('Dollar Value of Assets')
plt.xlabel('Days')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
# Comparison of different strategies
plt.plot(ucr_dollars_over_time, label='UCRP')
plt.plot(ubah_dollars_over_time, label='UBAH')
plt.plot(scurp_2_dollars_over_time, label='SCURP2')
plt.plot(scurp_4_dollars_over_time, label='SCURP4')
plt.plot(scurp_6_dollars_over_time, label='SCURP6')
plt.ylabel('Dollar Value of Assets')
plt.xlabel('Days')
plt.title('Comparison of Portfolio Strategies')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
# Semi-Constant Uniform Rebalancing Portfolio

# Rebalance every other day
scurp_2_dollars_over_time = get_float_array_from_file('../results/semi_const_rebalancing_dollars_over_time_interval_2.txt')

# Rebalance every 4th day
scurp_4_dollars_over_time = get_float_array_from_file('../results/semi_const_rebalancing_dollars_over_time_interval_4.txt')

# Rebalance every 6th day
scurp_6_dollars_over_time = get_float_array_from_file('../results/semi_const_rebalancing_dollars_over_time_interval_6.txt')

low_cost = get_float_array_from_file('../results_low_trans_cost/semi_const_rebalancing_dollars_over_time_interval_4.txt')
low_cost_2 = get_float_array_from_file('../results_low_trans_cost/2_semi_const_rebalancing_dollars_over_time_interval_4.txt')


plt.plot(scurp_2_dollars_over_time, label='SCURP2')
plt.plot(scurp_4_dollars_over_time, label='SCURP4')
plt.plot(scurp_6_dollars_over_time, 'b', label='SCURP6')
plt.plot(low_cost, 'r', label='SCURP4-low')
plt.plot(low_cost_2, 'g', label='SCURP4-lower')
plt.ylabel('Dollar Value of Assets')
plt.xlabel('Days')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()

# Conclusion: SCURP 4 doesn't perform better just b/c of transaction costs. It's actually better b/c
# of the nature of mean reversion not being daily! This is why RMR should work better!!
```
